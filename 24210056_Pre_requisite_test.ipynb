```python
# Q1
nat_num = [i for i in range(1, 21)]
print(f'List of natural numbers from 1 to 20: {nat_num}')
# Sum of even numbers
even_num_sum = 0
for num in nat_num:
    if num % 2 == 0:
        even_num_sum += num
print(f'Sum of even numbers from 1 to 20: {even_num_sum}')
# Product of odd numbers
odd_num_prod = 0
for num in nat_num:
    if num % 2 != 0:
        odd_num_prod *= num
print(f'Product of odd numbers from 1 to 20: {odd_num_prod}')
# Mean
total_sum = sum(nat_num)
mean = total_sum/len(nat_num)
print(f'Mean of natural numbers from 1 to 20: {mean}')
# Median
sorted_nat_num = sorted(nat_num)
if len(nat_num) % 2 == 0:
    median = (sorted_nat_num[len(nat_num)//2 - 1] + sorted_nat_num[len(nat_num)//2])/2
else:
    median = sorted_nat_num[len(nat_num)//2]
print(f'Median of natural numbers from 1 to 20: {median}')
# Standard Deviation
mean_diff_sq = [(i-mean)**2 for i in nat_num]
std = (sum(mean_diff_sq)/len(nat_num))**0.5
print(f'Standard Deviation of natural numbers from 1 to 20: {std}')
```

```python
# Q3
import matplotlib.pyplot as plt

def fib_seq(n):
    if n <= 0:
        return [] 
    if n == 1:
        return [0] 
    if n == 2:
        return [0, 1]  
    fib_numbers = [0, 1]
    for _ in range(2, n):
        next_number = fib_numbers[-1] + fib_numbers[-2]
        fib_numbers.append(next_number)
    return fib_numbers

# Plot of Fibonacci sequence
n=15
fibonacci_sequence = fib_seq(n)
plt.figure(figsize=(10, 6))
plt.plot(range(1, n + 1), fibonacci_sequence, marker='o', linestyle='-', color='g', label='Fibonacci Sequence')
plt.title('Fibonacci Sequence', fontsize=16)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Fibonacci Number', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.show()

```

```python
# Q9

def merge_dicts(dict1, dict2):
    merged_dict = dict1.copy()
    for k, v in dict2.items():
        if k in merged_dict:
            merged_dict[k] += v
        else:
            merged_dict[k] = v
    return merged_dict

dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
result = merge_dicts(dict1, dict2)
print(result)
```

```python
# Q10

def char_freq(string):
    string = string.lower()
    freq = {}
    for char in string:
        if char.isalpha():
            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1
    return freq

input_string = "This is sample text to find the count of each character in the string."
frequency = char_freq(input_string)
print(frequency)
characters = list(frequency.keys())
counts = list(frequency.values())
plt.figure(figsize=(14, 6))
colors = plt.cm.viridis(range(len(students)))
bars = plt.bar(characters, counts, color=colors, edgecolor='black')
plt.title('Character Frequency Distribution', fontsize=16)
plt.xlabel('Character', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, axis='y')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.1, str(height),
                ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
# Q11

def remove_duplicates(ip_lst):
    final_list = []
    uniq_elems = set()
    for i in ip_lst:
        if i not in uniq_elems:
            final_list.append(i)
            uniq_elems.add(i)
    return final_list

input_list = [10, 20, 10, 30, 20, 40, 30]
unique_elements = remove_duplicates(input_list)

_mean = sum(unique_elements)/len(unique_elements)

sorted_elems = sorted(unique_elements)
if len(sorted_elems) % 2 == 0:
    _median = (sorted_lst[len(sorted_elems) // 2] + sorted_lst[len(sorted_elems) // 2 - 1]) / 2
else:
    _median = sorted_lst[len(sorted_elems) // 2]
    
print("Original list:", input_list)
print("Unique elements:", unique_elements)
print("Mean of unique elements:", _mean)
print("Median of unique elements:", _median)
```

```python
# Q12

def string_analysis(strings):
    longest_string = max(strings, key=len)
    vowels = ['a','e','i','o','u','A','E','I','O','U']
    vowel_count = 0
    for s in strings:
        if s[0] in vowels:
            vowel_count += 1
    string_lengths = [len(s) for s in strings]
    return longest_string, vowel_count, string_lengths

strings = ["charan", "arun", "arti", "cinderella", "kaushik", "ironman", "peter parkar"]
longest_string, count_vowels, string_lengths = string_analysis(strings)
plt.figure(figsize=(10, 6))
plt.hist(string_lengths, bins=range(min(string_lengths), max(string_lengths) + 2), edgecolor='black', color='lightgreen')
plt.title('Distribution of String Lengths', fontsize=16)
plt.xlabel('String Length', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, axis='y')
plt.show()
print("Longest string:", longest_string)
print("Count of strings starting with a vowel:", count_vowels)
```

```python
# Q13
def sort_dicts_by_key(data, key):
    # Bubble Sort
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j][key] > data[j+1][key]:
                data[j], data[j+1] = data[j+1], data[j]
    return data

data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
sorted_data = sort_dicts_by_key(data, 'age')
print(sorted_data)
```

```python
# Q14

def word_filtering(input_string):
    words = input_string.split()
    # Remove words that are shorter than 4 characters
    filtered_words = []
    for word in words:
        if len(word) >= 4:
            filtered_words.append(word)
    filtered_words.sort()
    return filtered_words

input_string = "The Design of Approximation Algorithms by David P. Williamson and David B. Shmoy"
filtered_words = word_filtering(input_string)
word_lengths = [len(word) for word in filtered_words]
plt.figure(figsize=(10, 6))
bars = plt.bar(filtered_words, word_lengths, color='maroon', edgecolor='black')
plt.title('Word Lengths Distribution', fontsize=16)
plt.xlabel('Words', fontsize=14)
plt.ylabel('Length of Words', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True, axis='y')
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.1, str(height),
             ha='center', va='bottom', fontsize=12)
plt.show()
print("Processed words:", filtered_words)
```

```python
# Q15

def find_anagrams(word, words):
    sorted_word = sorted(word)
    anagrams = []
    for i in words:
        if sorted(i) == sorted_word:
            anagrams.append(i)
    return anagrams

word = "listen"
words = ["enlist", "google", "inlets", "banana"]
anagrams = find_anagrams(word, words)
print("Anagrams of '{}':".format(word), anagrams)
```

```python
# Q18

import re

def count_words(filename):
    with open(filename, 'r') as file:
        text = file.read().lower()  
        words = re.findall(r'\b\w+\b', text)
        word_count = {}
        for word in words:
            if word in word_count:
                word_count[word] += 1
            else:
                word_count[word] = 1
        
        total_words = sum(word_count.values())
        return total_words, word_count

def find_most_frequent_word(word_count):
    if word_count:
        most_frequent_word = max(word_count.items(), key=lambda x: x[1])
        return most_frequent_word
    else:
        return None


filename = r"D:\IITGN\Machine_Learning\words.txt"
total_words, word_count = count_words(filename)
if total_words > 0:
    print(f"Total number of words: {total_words}")
    # Find the most frequent word
    most_frequent = find_most_frequent_word(word_count)
    if most_frequent:
        print(f"The most frequent word is '{most_frequent[0]}' which occurred {most_frequent[1]} times.")
    # Visualization of word frequencies
    sorted_word_count = sorted(word_count.items(), key=lambda x: x[1], reverse=True)[:10]
    words, frequencies = zip(*sorted_word_count)
    # plot
    plt.figure(figsize=(12, 6))
    plt.bar(words, frequencies, color='lightgreen', edgecolor='black')
    plt.title('Top 10 Word Frequencies', fontsize=16)
    plt.xlabel('Words', fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(rotation=45, ha='right')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    for i, freq in enumerate(frequencies):
        plt.text(i, freq + 0.5, str(freq), ha='center', fontsize=10)

    plt.tight_layout()
    plt.show()
else:
    print("No data to process.")
```

```python
# Q19

import random

def gen_random_int(file_name, count=100, l=1, u=50):
    with open(file_name, 'w') as file:
        for _ in range(count):
            file.write(f"{random.randint(l, u)}\n")

def compute_mean(file_name):
    with open(file_name, 'r') as file:
        numbers = [int(line.strip()) for line in file]
    mean_value = sum(numbers) / len(numbers)
    return numbers, mean_value


file_name = "random_numbers.txt"
gen_random_int(file_name)
numbers, mean_value = compute_mean(file_name)
print(f"Mean of the numbers: {mean_value:.2f}")
plt.figure(figsize=(10, 6))
plt.hist(numbers, bins=range(min(numbers), max(numbers) + 2), color='skyblue', edgecolor='black', align='left')
plt.title('Distribution of Random Numbers', fontsize=16)
plt.xlabel('Numbers', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q21

def calculate_statistics(numbers):
    n = len(numbers)
    
    # Mean
    mean = sum(numbers) / n
    
    # Median
    sorted_numbers = sorted(numbers)
    if n % 2 == 0:
        median = (sorted_numbers[n // 2] + sorted_numbers[n // 2 - 1]) / 2
    else:
        median = sorted_numbers[n // 2]
    
    # Mode
    frequency = {}
    for num in numbers:
        if num in frequency:
            frequency[num] += 1
        else:
            frequency[num] = 1
    
    max_freq = max(frequency.values())
    mode = [key for key, value in frequency.items() if value == max_freq]
    
    # Standard Deviation
    variance = sum((x - mean) ** 2 for x in numbers) / (n - 1)
    std_dev = variance**0.5
    
    return mean, median, mode, std_dev



numbers = [1,1,1,2,2,7,7,9,3,3,3,3,3,6,6,6] 
mean, median, mode, std_dev = calculate_statistics(numbers)

print(f"Mean: {mean_value:.2f}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Standard Deviation: {std_dev:.2f}")
# box plot
plt.figure(figsize=(8, 6))
plt.boxplot(numbers, patch_artist=True, boxprops=dict(facecolor='skyblue', color='black'))
plt.title('Box Plot of Data', fontsize=16)
plt.ylabel('Values', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Q22

def normalize(numbers):
    min_val = min(numbers)
    max_val = max(numbers)
    return [(num - min_val) / (max_val - min_val) for num in numbers]

numbers = [10000, 222222, 3456, 4452, 5000, 6446, 777, 6562, 11111, 99999]
normalized_numbers = normalize(numbers)

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(numbers) + 1), numbers, label='Original Data', marker='o', linestyle='--')
plt.plot(range(1, len(normalized_numbers) + 1), normalized_numbers, label='Normalized Data', marker='o', linestyle='-', color='green')
plt.title('Original vs Normalized Data', fontsize=16)
plt.xlabel('Index', fontsize=14)
plt.ylabel('Value', fontsize=14)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
